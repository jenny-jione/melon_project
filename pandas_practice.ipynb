```python
import pandas as pd
FILENAME = "csv/Monthly_Top_10__2016-2022.csv"
df = pd.read_csv(FILENAME)
```

```python
df
```

```python
df['rank'] = df['rank'].fillna(-1).astype(int).replace({-1:None})
df
```

```python
type(df)
```

```python
# rank의 값이 1인 것만 추출
is_top = df['rank'] == 1
first_song = df[is_top]
first_song[:30]
```

```python
# year이 2022인 것만 추출
is_2022 = df['year'] == 2022
this_year = df[is_2022]
this_year
```

```python
# rank==1이면서 year==2022인 것만 추출
is_top = df['rank'] == 1
is_2022 = df['year'] == 2022
this_year_top_song = df[is_top & is_2022]
this_year_top_song
```

```python
def get_monthly_top_song(year: int):
    is_top = df['rank'] == 1
    is_thisyear = df['year'] == year
    return df[is_top & is_thisyear]
```
